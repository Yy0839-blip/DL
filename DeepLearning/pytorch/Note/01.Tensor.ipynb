-자동 미분:Tensor는 자동 미분(AutoGrad) 기능을 지원하여 딥러닝 모델의 학습에 필요한 기울기 계산을 자동으로 수행한다.:기울기(sgb)
#gpu(메모리와 연결하지 않음. 메모리를 가지고 있음. 할당을 해줘야함)-cpu+memory/hard

#DataSet(target&feature)과 DataLoader
-데이터의 전처리와 모델학습을 효율적으로 수행하기 위한 핵심 구성요소
-DataSet은 데이터와 라벨을 저장
-DataLoader는 이러한 데이터를 쉽게 순회 할수 있게 한다.
#epoch =>과대적합, 역전파시 순회(shuffle):골고루 학습:과대적합 방지

#DataSet과 DataLoader의 개념(역할)
#DataSet
-역할:데이터 샘플과 그에 해당하는 레이블을 저장
-주요 Method
 -init:데이터셋 초기화 및 데이터 로드
 -len:데이터셋의 샘플 갯수를 반환
 -getitem:주어진 인덱스에 해당하는 샘플을 반환(return)

#DataLoader
-역할:DataSet을 Batch단위로 쉽게 접근할수 있도록 하는 iterable 객체를 제공#연속=리스트
-주요 Method
 -batch_size:한번에 불러올 데이터의 크기
 -shuffle:데이터를 무작위로 섞을지 여부
 -num_workers:데이터를 로드할때 사용할 서브 프로세스의 수(LMM)

```python
#기본 데이터 셋 클래스 정의
from torch.utils.data import Dataset#기본 클래스
class SimpleDataset(Dataset):#상속:메소드=클라스 안 함수
    def __init__(self,data,labels):#생성자
        self.data=data
        self.labels=labels

    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):#__시스템에서 사용하는 변수
        return self.data[idx],self.labels[idx]
```

```python
data=[[1,2],
      [3,4],
      [5,6]]
labels=[0,1,0]

dataset=SimpleDataset(data,labels)
```

```python
print(dataset.data)
print(dataset.labels)
print(len(dataset))#인스턴스=객체, 리스트 아님
print(dataset[1])#get_item
```

```python
#기본 dataloader 사용:Shuffle하지 않고 전체를 한번에 사용하자 batch_size=3
from torch.utils.data import DataLoader
dataloader=DataLoader(dataset,batch_size=3,shuffle=False)
dataloader

for batch in dataloader:
    print(batch)#열중심&텐서로 바뀜
```

```python
#shuffle하고 2개씩 불러오기
dataloader=DataLoader(dataset,batch_size=2,shuffle=True)#true많이 씀
for batch in dataloader:
    print(batch)#2개=>나머지1개
```
