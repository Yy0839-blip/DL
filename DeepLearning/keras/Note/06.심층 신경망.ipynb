```python
from tensorflow import keras
from tensorflow.keras.layers import Input
(train_input, train_target),(test_input,test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
#train->train,valid 분리
from sklearn.model_selection import train_test_split#이미지 무조건 스케일링
train_scaled = train_input/255.0
train_scaled= train_scaled.reshape(-1,28*28)#컬럼 28*28

train_scaled, val_scaled,train_target, val_target= train_test_split(
    train_scaled,train_target,test_size=0.2,random_state=42
)
```

-입력층 : 배열의 갯수가 28*28=784(노드)개의 입력층 #활성화 함수=>층
-은닉층 : 갯수는 정해져 있지 않으나 , 출력층 보다는 많아야 한다.(2의 배수),Hyper parameter
:홀수 편가르기/입력층보다는 적어도 됨
-출력층 : 분류인 경우, 다중 분류는 softmax, 이진분류는 sigmoid

### 딥러닝 구성
#### 딥러닝 작업 흐름
# 1. 입력층(inputLayer)
-입력 데이터를 받는 층입니다. 데이터는 숫자로 변환된 상태로 들어옵니다.
-입력층은 데이터의 특징(feature)이나 변수를 신경망에 전달합니다.

#2.은닉층(Hidden Layer)
-입력층에서 받은 데이터를 처리하는 중간단계입니다.
하나 이상의 은닉층이 있을수 있습니다.(과대적합??)
-각 은닉층에서 활성화함수(Activation Function)가 적용됩니다.
-이 함수는 각 노드에서 계산된 값을 비선형적으로 변환하여(확률값) 다음 층으로 전달합니다.
(확률계산:시그모이드+a:0-1사이의 값)
-대표적인 활성화 함수는 Relu, Sigmoid, Tahn등이 있습니다.

#3.출력층(output Layer)
-은닉층에서 전달된 데이터를 최종적으로 처리하여 결과를 내는 층입니다.
-분류문제라면 softmax 같은 활성화 함수가 사용되어 예측확률을 계산합니다.

#4.손실함수(Loss Function)
-출력층에서 나온 결과와 실제값(정답)을 비교하여 오류를 계산합니다.
-손실함수는 예측이 얼마나 잘못되었는지 평가합니다.
-대표적인 손실함수로는 회귀문제에 사용되는 MSE(Mean squred Error):평균__오차와 분류문제에 사용되는 Cross_Entropy(교차 엔트로피) 등이 있습니다.

#5.최적화 함수(Optimization Function)
-손실함수에서 계산된 오류를 바탕으로 네트워크의 가중치를 업데이트 합니다.
이를 통해 모델이 더 좋은 예측을 할수 있도록 학습이 진행됩니다.
-대표적인 최적화 함수로는 확률적 경사하강법(Stochastic Gradient:기울기 Descent:SGD)이나 Adam 등이 있습니다.
-최적화 함수는 역전파(back propagation)과정에서 가중치와 편향을 조정하는 역할을 합니다.

이 과정이 여러번 반복되면서 신경망은 학습을 통해 점점 더 좋은 결과를 내놓게 됩니다.
