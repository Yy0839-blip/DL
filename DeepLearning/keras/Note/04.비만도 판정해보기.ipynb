#data 정규화

```python
#minmax 차이 많이남 정규화
#weight, height<-des
csv['weight']=csv['weight']/csv['weight'].max()
csv['height']=csv['height']/csv['height'].max()

```

```python
csv.head()#label=category 
```

```python
#label 원핫인코딩(딕셔너리:key/value)
bclass={
    'thin':[1,0,0],
    'normal':[0,1,0],
    'fat':[0,0,1]
}
y=[]#if 필요없음
for v in csv['label']:
    #print(v)
    y.append(bclass[v])

y
```

---
###훈련용과 테스트

```python
#Train
x_train=csv[['height','weight']][:15000]
y_train=y[:15000]
```

```python
#Test Data
x_test=csv[['height','weight']][15000:]
y_test=y[15000:]
```

#### Deep Learning Model 만들기


```python
#입력층:2
#은닉층:512, activation='relu'
#출력층:3, activation='softmax'#<<<<<<타겟?의수
#loss:optimizer='rmsprop',loss='categorical_crossentropy'
from tensorflow import keras
from tensorflow.keras.layers import Input

model=keras.Sequential()#히든 레이어 만들기
model.add(Input(shape=(2,)))#튜플
# 첫 번째 은닉층
model.add(keras.layers.Dense(512, activation='relu'))

# 두 번째 은닉층 추가 
model.add(keras.layers.Dense(256, activation='relu'))

# model.add(keras.layers.Dense(512,activation='relu'))
model.add(keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',#손실 함수   
    metrics=['accuracy']
)#오차계산
```

```python
csv[['height','weight']]#두번쨰 은닉층은 갯수는 첫번째 은닉층 보다 작아야함
```

```python
import numpy as np#시그모이드 2개만 가능
#verbose=0은 학습 과정 출력 여부, 0은 출력하지 않음, 1은 진행 바 형태로 출력, 2는 한 줄에 한 epoch마다 출력
```

```python
model.fit(np.array(x_train), np.array(y_train), epochs=100)
model.evaluate(np.array(x_test), np.array(y_test))
```
