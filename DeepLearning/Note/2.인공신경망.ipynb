```python
#학습전 균등하게 확인 몇개씩 있는지
#train_target의 종류 및 갯수 파악
import numpy as np
print(np.unique(train_target, return_counts=True))#데이터 많을떄 넘파이/60000개씩 10종류?
```

#로지스틱 회귀로 패션 아이템 분류

```python
#데이터 정규화 (이미지 최대값255) 및 2차 행렬로 만들기
train_scaled = train_input / 255.0
train_scaled=train_scaled.reshape(-1, 28*28)#로우는 너가 알아서
```

```python
#dimension 확인(그림:2차원)
train_scaled.shape
```

```python
#로지스틱 회귀 모델
from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier#10종류

sc = SGDClassifier(loss='log_loss', max_iter=5, random_state=42)#로지스틱회귀/최대반복횟수/랜덤시드
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1)#교차검증/병렬처리
print("평균:",np.mean(scores['test_score']))#평균점수   
#에러뜨고 선생님과 자릿수가 다름
```

#인공신경망(히든레이어 하나, 리스트?와 구조 비슷)
#keras model 만들기

```python
#Train data를 훈련과 검증 데이터로 나누기
from sklearn.model_selection import train_test_split
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)#검증데이터 20%/랜덤시드
```

```python
#dimension확인
print(train_scaled.shape, train_target.shape)
print(val_scaled.shape, val_target.shape)
```

```python
#구조를 같게 하기 위해서 입력층과 출력층 만들기 / keras 손실함수 : 매개변수 조정? activation: 활성화함수 |y-y^|
```

#keras로 층 만들기

```python
from tensorflow.keras.layers import Input

model=keras.Sequential()#순차적 모델 #컬럼28*28=784
model.add(Input(shape=(28*28,)))#입력층 추가 튜플구조(피처를 세로로 세움)
model.add(keras.layers.Dense(10, activation='softmax'))#출력층/10개로 분류/확률값으로 출력
# model.summary()#구조확인 #출력층 10개= 클라스 수 더 높은 것 맞추기/분류 여러개 sofrtmax/두개 sigmoid  
# 활성화 함수만 적음 
```

```python
#손실함수#분류
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])#손실함수/평가기준#정수 sparse/target data 원핫 인코딩
```

```python
#모델훈련#지도학습#max=5
model.fit(train_scaled, train_target, epochs=5)#훈련/에포크:전체 데이터 1회 반복/5회 반복
```

```python
#검증 데이터 확인 훈련이 끝나면 검증데이터로 평가
model.evaluate(val_scaled, val_target)#적절한 과적합
```
