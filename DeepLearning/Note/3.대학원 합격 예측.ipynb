#DeepLearningModelл§Ңл“Өкё°

```python
from tensorflow import keras
from tensorflow.keras.layers import Input

#мһ…л Ҙмёө : 3к°ң(к·ёл Ҳ, gpa, rank)
#мқҖлӢүмёө : 32:н•ҳмқҙнҚјнҢҢлқјлҜён„°
#м¶ңл Ҙмёө : 1к°ң(н•©кІ© м—¬л¶Җ)<<0кіј 1лЎң н‘ңнҳ„/true/falseлЎң н‘ңнҳ„
#нҷңм„ұнҷ” н•ЁмҲҳ : sigmoid(0кіј 1 мӮ¬мқҙмқҳ к°’мқ„ м¶ңл Ҙн•ҳлҠ” н•ЁмҲҳ)/tanh(-1кіј 1 мӮ¬мқҙмқҳ к°’мқ„ м¶ңл Ҙн•ҳлҠ” н•ЁмҲҳ)/relu(0ліҙлӢӨ мһ‘мқҖ к°’мқҖ 0мңјлЎң, 0ліҙлӢӨ нҒ° к°’мқҖ к·ёлҢҖлЎң м¶ңл Ҙн•ҳлҠ” н•ЁмҲҳ)/
#softmax(м¶ңл Ҙк°’мқҳ н•©мқҙ 1мқҙ лҗҳлҸ„лЎқ н•ҳлҠ” н•ЁмҲҳ, лӢӨмӨ‘ нҒҙлһҳмҠӨ л¶„лҘҳм—җм„ң мӮ¬мҡ©)/leaky relu(0ліҙлӢӨ мһ‘мқҖ к°’мқҖ мһ‘мқҖ мқҢмҲҳлЎң, 0ліҙлӢӨ нҒ° к°’мқҖ к·ёлҢҖлЎң м¶ңл Ҙн•ҳлҠ” н•ЁмҲҳ)/elu(0ліҙлӢӨ мһ‘мқҖ к°’мқҖ м§ҖмҲҳ н•ЁмҲҳлЎң, 0ліҙлӢӨ нҒ° к°’мқҖ к·ёлҢҖлЎң м¶ңл Ҙн•ҳлҠ” н•ЁмҲҳ)

model=keras.Sequential()#нһҲл“  л Ҳмқҙм–ҙ л§Ңл“Өкё°
model.add(Input(shape=(3,)))#нҠңн”Ң
model.add(keras.layers.Dense(32,activation='tanh'))#мҲ«мһҗкі„мӮ°? #мқҖлӢүмёө:2мқҳ л°°мҲҳ(мһ…л ҘмёөліҙлӢӨ л§Һм•„м•јн•Ё)
model.add(keras.layers.Dense(1,activation='sigmoid'))#1
#model.add(keras.layers.Dense(1,activation='softmax'))#2
```

```python
model.compile(
    optimizer='adam',#мөңм Ғнҷ” м•Ңкі лҰ¬мҰҳ #н•ҳмқҙнҚјнҢҢлқјлҜён„°
    loss='binary_crossentropy',#мҶҗмӢӨ н•ЁмҲҳ   
    metrics=['accuracy']
)#мҳӨм°Ёкі„мӮ°
```

#optimizer : лӘЁлҚёмқҳ к°ҖмӨ‘м№ҳлҘј м—…лҚ°мқҙнҠён•ҳлҠ” л°©лІ•мқ„ кІ°м •н•ҳлҠ” м•Ңкі лҰ¬мҰҳ:learning rate(н•ҷмҠөлҘ )мқ„ м•Ңл§һкІҢ мҲҳм •н•ҙмӨҢ
#:adam,adagrad,rmsprop,sgd,adadelta л“ұ лӢӨм–‘н•ң мөңм Ғнҷ” м•Ңкі лҰ¬мҰҳмқҙ мЎҙмһ¬

#loss : лӘЁлҚёмқҳ мҳҲмёЎк°’кіј мӢӨм ңк°’мқҳ м°ЁмқҙлҘј кі„мӮ°н•ҳлҠ” н•ЁмҲҳ
#:мҶҗмӢӨн•ЁмҲҳ:binary_crossentropy(мқҙм§„ л¶„лҘҳм—җм„ң мӮ¬мҡ©),categorical_crossentropy(лӢӨмӨ‘ нҒҙлһҳмҠӨ л¶„лҘҳм—җм„ң мӮ¬мҡ©),mean_squared_error(нҡҢк·Җ л¬ём ңм—җм„ң мӮ¬мҡ©) л“ұ лӢӨм–‘н•ң мҶҗмӢӨ н•ЁмҲҳк°Җ мЎҙмһ¬

#metrics : лӘЁлҚёмқҳ м„ұлҠҘмқ„ нҸүк°Җн•ҳлҠ” м§Җн‘ң  
#:accuracy(м •нҷ•лҸ„),precision(м •л°ҖлҸ„),recall(мһ¬нҳ„мңЁ),f1_score(f1 м җмҲҳ) л“ұ лӢӨм–‘н•ң нҸүк°Җ м§Җн‘ңк°Җ мЎҙмһ¬

```python
#н•ҷмҠөмӢңнӮӨкё°
x
# л”Ҙлҹ¬лӢқ м–ҙл Ҳмқҙ
```

```python
import numpy as np
#verbose=0мқҖ н•ҷмҠө кіјм • м¶ңл Ҙ м—¬л¶Җ, 0мқҖ м¶ңл Ҙн•ҳм§Җ м•ҠмқҢ, 1мқҖ м§„н–ү л°” нҳ•нғңлЎң м¶ңл Ҙ, 2лҠ” н•ң мӨ„м—җ н•ң epochл§ҲлӢӨ м¶ңл Ҙ
model.fit(np.array(x),np.array(y),epochs=1000)#н•ҷмҠөмӢңнӮӨкё°, epochsлҠ” н•ҷмҠө нҡҹмҲҳ, verboseлҠ” н•ҷмҠө кіјм • м¶ңл Ҙ м—¬л¶Җ
#л§Һмқҙ л°ҳліө epochs=1000<-100
```

```python
model.evaluate(np.array(x),np.array(y))
```
