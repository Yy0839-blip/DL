#딥러닝이란
-딥러닝은 심층신경망(Deep Neural Network:DNN)이라고 불리운다.
-뉴런(노드)으로 구성된 레이어를 여러개 연결해서 구성한 네트워크(설계도:ERD)
-대표적인 예로 cnn(convolution neural network:사물을 이해하는 과정):얼굴인식,
RNN(recurrent neural network:문맥을 이해하는 과정):뺑뺑 돈다.:ADP시험에 나옴


#딥러닝의 학습
-학습목표: 모델에 입력값을 넣었을떄 출력값이 최대한 정답값과 일치하게 하는것.
학습: 최초 딥러닝 모델에 매개변수(weight,bias)를 무작위로 부여한후,:코딩사이트?
반복학습을 통해서 모델의 출력값이 최대한 정답과 일치하게 매개변수를 조정하는 과정
(오차를 줄여 : 근삿값)

#순전파(forward propagation)
빨간점:활성함수?한번에 작업이 끝남
-왼쪽에서 오른쪽으로 흘러가는 과정
-순전파에 의해 딥러닝의 출력값(y^:예측값)이 결정, 정답은 (y)

#손실함수 (loss function 오차를 해결):|y-y^|
-출력값과 정답의 차이
-출력값과 정답이 일치할수록 손실함수의 값은 작아진다.
-보통 회귀에는 평균제곱오차(Mean Squared Error:MSE),:음수 없애기 위해 제곱
-분류 문제에는 크로스 엔트로피(cross Entropy) 를 사용
-매개변수(weight,Bias)를 조절해서 손실함수의 값을 최저로 만드는 과정을 최적화(Optimization)이라 한다.
-최적화 작업은 optimizer을 통해 이뤄지며 Optimizer는 역전파(back propagation)과정을 수행해서 딥러닝 모델의 <매개변수>를 최적화한다.
(에포크)

#최적화(optimization):w와 b를 결정하는 과정
-최적화의 대표적인 방법은 경사하강법(batch gradient descent)
-반복적으로 손실함수에 대한 모델 매개변수의 미분값(기울기)을 구한후 그 미분(0:선->점,선모양이 달라짐)값의 반대방향으로 <매개변수>를 조절해 나가면 결국레는 최저 손실함수값(제일 꼭지점)에 도달한다.

#역전파(back propagation):순정파의 거꾸로:다시 작업
-optimizer는 손실함수의 값을 최소화 하기 위해 역전파를 사용해 딥러닝 모델의 모든 <매개변수>를 변경한다. variable x->parameter:function

#학습률(learning Rate):얼마나 이동하는지/적으면 금방 맞춤(딥러닝 데이터 많아서 밸리드 검증 안함)
-학습률을 크게 설정하면 매개변수 변경치가 커져서 정답으로 수렴하지 않을수도 있고, 너무 작으면 학습 시간이 상당히 오래 걸릴수 있다.(adam/과대적합?)
